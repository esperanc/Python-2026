## Utilitário para testes no Beecrowd

Às vezes, testar a solução pode requerer digitar os dados de teste várias vezes. 

Para evitar entrar manualmente com os dados para cada teste, você pode usar a seguinte função:

>`run_from_string (function, string)`
>
Esta função roda `function` tomando como entrada as linhas de `string` ao invés de solicitar os dados do teclado diretamente. 

Assim você pode testar automaticamente seu código para os mesmos dados sem precisar digitá-los a cada rodada.


```python
def run_from_string (function, string):

    import builtins

    lines = iter(string.splitlines())
    
    def fake_input(prompt=""):
        try:
            val = next(lines)
            #print(f"{prompt}{val}") 
            return val
        except StopIteration:
            raise EOFError("No more input data.")
            
    original_input = builtins.input
    builtins.input = fake_input

    function () 

    builtins.input = original_input
```

## Exemplo: Problema de Josephus

O Problema de Josephus em https://judge.beecrowd.com/pt/problems/view/1030 é aparentemente simples, 
mas uma solução usando um algoritmo trivial não vai ser aceita pelo juiz do Beecrowd.

A função `josephus` abaixo contém  uma possível solução para o problema, conforme discutido [neste artigo do Wikipedia](https://en.wikipedia.org/wiki/Josephus_problem).


```python
def josephus (n: int, k:int) -> int:
    """Resolve o problema de Josephus"""
    g = 0
    for N in range (2,n+1):
        g = (g+k) % N
    return g+1
    
def run_josephus():
    """Roda a função josephus para alguma massa de dados do beecrowd"""
    nc = int(input())
    for i in range(nc):
        nstr,kstr = input().split()
        n,k = int(nstr),int(kstr)
        result = josephus(n,k)
        print (f"Case {i+1}: {result}")
```

```python
## Dados do beecrowd. 
dados_josephus = """3
5 2
6 3
1234 233"""
```

```python
run_from_string(run_josephus, dados_josephus)
```
