```python
def poligono(lado : float, n : int) -> None:
    """
    Desenha um polígono no ambiente jpyturtle
    Args:
        n: número de lados
        lado: comprimento do lado
    """
    for i in range(n):
        forward(lado)
        left(360/n)
```

```python
def poligono_direito (lado,n):
    """
    Desenha um polígono no ambiente jpyturtle à direita da tartaruga
    Args:
        n: número de lados
        lado: comprimento do lado
    """
    for i in range(n):
        forward(lado)
        right(360/n)
```

```python
make_turtle(height=300)

poligono_direito (50,4)  
poligono_direito (60,5)
poligono_direito (70,6)
```

```python
def poligono_flor(lado:float, n : int, m: int) -> None:
    """
    Desenha um polígono duplo na forma de flor
    Args:
        lado: o comprimento do lado
        n: o número de lados do interior
        m: o número de lados de cada pétala
    """
    for i in range (n):
        poligono_direito(lado,m)
        forward(lado)
        left(360/n)
```

```python
make_turtle (height = 300, delay = 0)
poligono_flor (40, 7, 5)
```

```python
import math

def arco(r : float, angulo: float) -> None:
    """
    Desenha um arco de círculo à esquerda da tartaruga 
    Args
        r: raio do arco
        angulo: ângulo do arco
    """
    # Aproximação com n segmentos
    n = 20
    dang = angulo / n
    # (lado/2) / r = seno (dang/2)
    lado = math.sin (dang * math.pi / 180 / 2) * r * 2
    for i in range(n):
        left(dang/2)
        forward(lado)
        left(dang/2)
```

```python
make_turtle (height = 300, delay = 0.0)
arco (50, 180)
```

```python
def petala (r : float, angulo : float) -> None:
    """
    Desenha uma pétala com dois arcos
    Args:
        r : raio do arco
        angulo: ângulo do arco
    """
    right (angulo/2)
    arco (r, angulo)
    right (angulo)
    left(180)
    arco (r, angulo)
    right (angulo/2)
    left (180)
```

```python
make_turtle (height = 300, delay = 0.0)
petala (90, 90)
petala (80, 120)
```

```python
def flor (r : float, angulo : float, n : int) -> None:
    """
    Desenha uma flor com pétalas construidas com arcos
    Args:
        r : raio do arco
        angulo: ângulo do arco
        n : número de pétalas
    """
    dang = 360 / n
    for i in range (n):
        petala (r, angulo)
        left (dang)
```

```python
make_turtle (height = 300, delay = 0.0)
flor (50, 60, 5)
flor (70, 60, 5)
flor (50, 240, 5)
```

```python
def caixa(largura : float, altura: float) -> None:
    """
    Desenha uma caixa retangular centralizada na
    posição atual da tartaruga
    Args:
        largura: lado da caixa ao longo da direção atual
        altura: lado perpendicular à direção atual 
    """
    penup()
    forward(largura/2) 
    left(90)
    forward(altura/2)
    pendown()
    for i in range(2):
        left(90)
        forward(largura)
        left(90) 
        forward(altura)
    penup()
    back(altura/2)
    right(90)
    back(largura/2)
```

```python
make_turtle (height = 300, delay = 0.0)
caixa (200,150)
```

```python
def caixas_aninhadas(largura, altura, n):
    """
    Desenha caixas aninhadas, todas com a mesma razão de aspecto,
    sendo que cada caixa de dentro é orientada a 90 graus da caixa
    de fora
    Args: 
        largura: largura da caixa mais externa
        altura: altura da caixa mais externa
        n: número total de caixas
    """
    
    for i in range(n):
        caixa(largura,altura)
        if largura>altura:
            largura = altura**2 / largura
        else:
            altura = largura**2 /altura
```

```python
make_turtle (height = 300, delay = 0.0)
caixas_aninhadas (200,150,5)
```
