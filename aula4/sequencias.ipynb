```python
xpto = "abc", "def", "ghi"
com_tracos = () # tupla vazia
for s in xpto:
    r = "" # string vazia
    for c in s:
        r = r + c + "-"
    com_tracos = com_tracos + (r,) # adiciona r à tupla
com_tracos # Imprime ('a-b-c-', 'd-e-f-', 'g-h-i-')
```

```python
def insere (lista:list, posicao:int, valor) -> None:
    """Insere valor em lista[posicao]"""
    lista[posicao:posicao]=[valor]
```

```python
lista = [1,2,3]
insere(lista, 1, 4)
lista
```

```python
def insere_final(lista : list, valor) -> None:
    """Insere valor no final da lista"""
    insere(lista,len(lista), valor)
```

```python
lista = [1,2,3]
insere_final(lista,99)
lista
```

```python
def remove (lista:list, posicao = 0) -> None:
    """Remove de lista o elemento na posicao dada"""
    lista[posicao:posicao+1] = []
```

```python
lista = [1,2,3]
remove(lista,1)
lista
```

```python
def remove_final(lista : list) -> None:
    """Remove o último elemento da lista"""
    lista[-1:]=[]
```

```python
lista = [1,2,3]
remove_final(lista)
lista
```

```python
matriz = [[1,2,3],[4,5,6],[7,8,9]]

def coluna(m:list, i:int) -> list:
    """Retorna a coluna i da matriz m"""
    col = []
    for linha in m: 
        col.append(linha[i])
    return col

def transposta(m: list) -> list:
    """Retorna a transposta da matriz m"""
    tr = []
    for i in range(len(m[0])): # Varrer colunas de m
        tr.append(coluna(m,i))
    return tr
transposta(matriz)
            
```

```python
def prod_escalar(u : list, v:list) -> list:
    """Retorna o produto escalar de u e v"""
    soma = 0
    for i in range(len(u)): 
        soma += u[i]*v[i]
    return soma

def mult (a : list, b: list) -> list:
    m = []
    for i in range(len(a)): 
        linha = []
        for j in range(len(b[0])): 
            linha.append(prod_escalar(a[i],coluna(b,j)))
        m.append(linha)
    return m

mult([[1, 2], [3, 4], [5, 6]], [[1, 2], [3, 4]])
                         
```

```python
def primos(n:int) -> list:
    """Retorna todos os primos até n"""
    crivo = list(range(2,n))
    lista = []
    while len(crivo)>0:
        x = crivo[0]
        lista.append(x)
        crivo2 = []
        for y in crivo:
            if y%x != 0: 
                crivo2.append(y)
        crivo = crivo2
    return lista
primos(40)            
```

```python
def fibonacci(n:int) -> list:
    """Retorna os n primeiros números de Fibonacci"""
    fib = [0,1]
    if n<=2: return fib[:n]
    for i in range(2,n):
        fib.append(fib[i-1]+fib[i-2])
    return fib
fibonacci(10)
```
