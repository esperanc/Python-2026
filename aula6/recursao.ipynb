```python
def busca_binaria (lista : list, x)-> bool:
    def busca (i,j): # Está em lista[i:j]?
        if j-i == 0: return False
        if j-i == 1: return x == lista[i]
        meio = (i+j)//2
        if x < lista[meio]: return busca(i,meio)
        return busca(meio,j)
    return busca(0,len(lista))
```

```python
def bb_iterativa (lista: list, x) -> bool:
    i,j = 0, len(lista)
    while j-i > 1:
        meio = (i+j)//2
        if x < lista[meio]: j = meio
        else: i = meio
    if j-i == 0: return False
    return x == lista[i]
```

```python
def bb(lista:list, x)-> bool:
    if len(lista) == 0: return False
    if len(lista) == 1: return x == lista[0]
    meio = len(lista)//2
    if x < lista[meio]: return bb(lista[:meio],x)
    return bb(lista[meio:],x)
```

```python
def busca_abb(abb:list, x: int) -> bool:
    if len(abb) == 0: return False
    if abb[0] == x: return True
    if abb[0] > x: 
        return busca_abb(abb[1], x)
    return busca_abb(abb[2],x)
```

```python
busca_abb(abb,12)
```

```python
def insere_abb (abb, x):
    if len(abb)==0: abb[:] = [x,[],[]]
    elif x < abb[0]:
        insere_abb(abb[1],x)
    else:
        insere_abb(abb[2],x)
```

```python
def menor_abb(abb):
    """Retorna o menor nó da abb"""
    while len(abb[1])>0: abb = abb[1]
    return abb
```

```python
def remove_abb (abb, x):
    """Remove x da abb"""
    if len(abb)==0: return
    if abb[0]==x:
        if len(abb[1])==0:
            abb[:] = abb[2]
        elif len(abb[2])==0:
            abb[:] = abb[1]
        else:
            s = menor_abb(abb[2])
            abb[0]=s[0]
            remove_abb(s, s[0])
    elif x<abb[0]:
        remove_abb(abb[1],x)
    else:
        remove_abb(abb[2],x)
            
```

```python
def visita_em_ordem(abb, lista):
    if len(abb)>0:
        visita_em_ordem(abb[1],lista)
        lista[len(lista):]=[abb[0]]
        visita_em_ordem(abb[2],lista)
        
def ordena_com_abb(lista):
    abb = []
    for x in lista: 
        insere_abb(abb,x)
    lista = []
    visita_em_ordem(abb,lista)
    return lista
```
